```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def euclidean_distance(X1,Y1,X2,Y2):
    return math.sqrt(((X1-X2)**2) + ((Y1-Y2)**2))
```

```python
def Monte_Carlo_Simulation(trials,r,task):
    hit_x = []
    hit_y = []
    miss_x = []
    miss_y = []

    hit = 0

    for i in range(trials):
        X_rand = np.random.uniform(-8.5,8.5)
        Y_rand = np.random.uniform(0,8.5)

        if euclidean_distance(X_rand,Y_rand,0,0) <= r:
            hit+=1
            hit_x.append(X_rand)
            hit_y.append(Y_rand)
        else:
            miss_x.append(X_rand)
            miss_y.append(Y_rand)
    
    if task=="a":
        plt.scatter(hit_x,hit_y,color='green')
        plt.scatter(miss_x,miss_y,color='red')
        plt.title('graph for set-3')
        
    return (2*17*hit)/(8.5*trials)
```

```python
def Task_a(trials):
    r = 8.5

    plt.clf()
    # Drawing the half circle equation: y = SQRT(r^2 - x^2)
    x = np.arange(-8.5,8.5,0.0001)
    y = np.sqrt((r**2)-(x**2))
    plt.plot(x,x*0,color='black')
    plt.plot(x,y,color='blue')
    plt.axis('equal')

    print(f'Pi value : {Monte_Carlo_Simulation(trials,r,"a")}')
    plt.show() 
```

```python
def Task_b():
    x = []
    errors = []
    for i in range(10,1001,8):
        x.append(i)
        errors.append(abs(np.pi - Monte_Carlo_Simulation(trials=i,r=8.5,task="b")))
    plt.clf()
    plt.plot(x,[0]*len(x),color='black')
    plt.plot(x,errors)
    plt.title('error graph')
    plt.xlabel('trial no')
    plt.ylabel('error')
    plt.show()
```

```python
Task_a(int(input("Enter trial Number : ")))
```
